```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/machine_learning/over_fitting2.csv')

# split data
train_data = df[df['Process'] == 'Train']
validation_data = df[df['Process'] == 'Validation']

# filter data
train_jpara = train_data[train_data['Data'] == 'JParaCrawl']
train_jesc = train_data[train_data['Data'] == 'JESC']
val_jpara = validation_data[validation_data['Data'] == 'JParaCrawl']
val_jesc = validation_data[validation_data['Data'] == 'JESC']

# create x axis labels (Epoch-Chunk)
train_jpara['Epoch-Chunk'] = train_jpara['Epoch'].astype(str) + '-' + train_jpara['Chunk'].astype(str)
train_jesc['Epoch-Chunk'] = train_jesc['Epoch'].astype(str) + '-' + train_jesc['Chunk'].astype(str)
val_jpara['Epoch-Chunk'] = val_jpara['Epoch'].astype(str) + '-' + val_jpara['Chunk'].astype(str)
val_jesc['Epoch-Chunk'] = val_jesc['Epoch'].astype(str) + '-' + val_jesc['Chunk'].astype(str)



# create a graph
fig, ax1 = plt.subplots(figsize=(10, 6))

# left y axisTrain Loss
ax1.set_xlabel('Epoch-Chunk')
ax1.set_ylabel('Train Loss', color='tab:blue')

# JParaCrawl training loss
ax1.plot(train_jpara['Epoch-Chunk'], train_jpara['Loss'], label='Train Loss (JParaCrawl)', marker='o', color='tab:blue')

# JESC training loss
ax1.plot(train_jesc['Epoch-Chunk'], train_jesc['Loss'], label='Train Loss (JESC)', marker='o', color='tab:green')

ax1.tick_params(axis='y', labelcolor='tab:blue')

# flip the left y axisrepresent in a descending order
ax1.invert_yaxis()

# right x axisValidation Loss
ax2 = ax1.twinx()
ax2.set_ylabel('Validation Loss', color='tab:red')


# JParaCrawl validation loss
ax2.plot(val_jpara['Epoch-Chunk'], val_jpara['Loss'], label='Validation Loss (JParaCrawl)', marker='o', color='tab:red')

ax2.plot(val_jesc['Epoch-Chunk'], val_jesc['Loss'], label='Validation Loss (JESC)', marker='o', color='tab:purple')

ax2.tick_params(axis='y', labelcolor='tab:red')

# flip the right y axisrepresent in a descending order
ax2.invert_yaxis()

# flip the x axis
fig.autofmt_xdate(rotation=45)

# title of the graph
plt.title('Training and Validation Loss (JParaCrawl and JESC)')

# add legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# set a layout
fig.tight_layout()

plt.show()

```
