```python
# [:-2] is for deleting the last 2 elements "EOS" and "".
tagger = MeCab.Tagger(r'-d C:\\users\\hirom\\nlp\\llm\\cuda\\lib\\site-packages\\ipadic\\dicdir')
parsed_txt = tagger.parse(txt)
elements = parsed_txt.split('\n')[:-2]
elements
```

```python
# with dict class, words do not need to be wrapped with (), 
# and key and value are connected with = 
results = []

for element in elements:
    parts = element.split(',')
    surface_pos, pos1,base = parts[0], parts[1], parts[-3]
    surface, pos = surface_pos.split('\t')
    results.append(dict(表層形=surface, 基本形=base, 品詞=pos, 品詞1=pos1))
    
results
```

```python
result = results[0]
result
```

```python
# use "add" to add something to a set. if it's a list, use append.
surface_verbs = set()
for result in results:
    if result['品詞'] == '動詞':
        surface_verbs.add(result['表層形'])

surface_verbs
```

```python
# the same value is eliminated by set
base_verbs = set()
for result in results:
    if result['品詞'] == '動詞':
        base_verbs.add(result['基本形'])

base_verbs
```

```python
from collections import defaultdict
```

```python
# defaultdict is a dict. 
# Unlike normal dictionary, it creates key if the key doesn't exist and sets the value 0 by default.
# If it's a normal dictionary, if scentence to check whether there is the key is necessary
# and value has to be set to 0.
# if the value of '品詞' key is not '記号', its value of '基本形' is added.
word_freq = defaultdict(int)

for result in results:
    if result['品詞'] != '記号':
        word_freq[result['基本形']] += 1
word_freq
```

```python
# clustering: belongs to unsupervised learning
# cluster 1000 words with a method called Kmeans
# wv: word2vec
# use .keys() method for a dictionary model.wv.vocab

max_vocab = 1000

# retrieve 1000 words(key in a dictionary "model.key_to_index") and store them to vocab 
vocab = list(model.key_to_index.keys())[:max_vocab]

# get 1000 vectors for words and store them to vectors
vectors = [model[word] for word in vocab]
```

```python
# set the number of cluster to be 100

from sklearn.cluster import KMeans
from sklearn.utils import parallel_backend

n_clusters = 100

"""
Create a model instance.

The verbose parameter controls the verbosity of the output during the execution of the algorithm. 
It determines how much information is printed to the console while the algorithm runs.
verbose=1: Minimal output. Provides some information about the progress of the algorithm.

The random_state parameter is used to set the seed for the random number generator. 
Setting a seed ensures that the random processes in the algorithm produce the same results every time you run the code, which is crucial for reproducibility.
random_state=42: The number 42 is used as the seed for the random number generator. You can use any integer value here.

"""
kmeans_model = KMeans(n_clusters=n_clusters, verbose=1, random_state=42)

# Use parallel_backend to control the parallelism
# n_jobs parameter specifies the number of CPU cores to use for parallel processing.
# n_jobs=-1 means that all available CPU cores will be used for the computation.
# 'loky' is a backend used by the joblib library for parallel computing in Python. 
# joblib is a library used for running tasks in paralle.
# loky: The default backend for joblib. It is used for process-based parallelism and is robust.

with parallel_backend('loky', n_jobs=-1):
    # make the model learn vectors
    kmeans_model.fit(vectors)
```

```python
# label 1000 words with 100 different numbers (number of n_clusters)
cluster_labels = kmeans_model.labels_

# cluster words with label numbers and represent them in a dictionary in a list.{word:label}
# zip can retrieve one element in two lists(cluster_labels, vocab) at the same time.
# put cluster_id and word into a dictionary cluster_to_words
cluster_to_words = defaultdict(list)
for cluster_id, word in zip(cluster_labels, vocab):
    cluster_to_words[cluster_id].append(word)
```

```python
# Show the first 10 clusters (value) in the dictionary "cluster_to_eords". These do not line in the order of cluster_id
for words in cluster_to_words.values():
    print(words[:10])
```

```python
import bhtsne
import numpy as np
```

```python
# bhtsne: decrease dimensions
# decrease the dimensions of vectors from 300 to 2 to illustrate in a graph.
# The rand_seed parameter is used to set the seed for the random number generator to produce the same results every time you run the code.You can use any integer value here.
# 1000 vectors * 2 dimensions
embedded = bhtsne.tsne(np.array(vectors).astype(np.float64), dimensions=2, rand_seed=123)
```

```python
embedded.shape
```

```python
# T: transpose: flip row and column (転置)
embedded.T.shape
```

```python
# The index 0 dimension of all the 1000 words 
embedded.T[0]
```

```python
# set the size of the graph
plt.figure(figsize=(20, 20))

# x axis: the first dimension of 1000 words vectors, y axis: the second dimension of 1000 words vectors
plt.scatter(np.array(embedded).T[0], np.array(embedded).T[1])
# embedded: 2 dimension vectors, vocab: words
# set embedded to (x, y) and vocab to name
# show names for value of each vector with annotate
for (x, y), name in zip(embedded, vocab):
    plt.annotate(name, (x, y))
plt.show()
```
